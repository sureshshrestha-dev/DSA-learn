```python
arr = [10, 20, 30, 40]

# Access
print(arr[1])  # O(1)

# Insert at end
arr.append(50)  # O(1)

# Insert at position
arr.insert(2, 25)  # O(n)

# Delete
arr.remove(30)  # O(n)

# Traverse
for x in arr:  # O(n)
    print(x)
```

| Operation       | Time Complexity |
| --------------- | --------------- |
| Access by index | O(1)            |
| Append          | O(1)            |
| Insert middle   | O(n)            |
| Delete          | O(n)            |
| Search          | O(n)            |


Key Idea: Arrays are fast for index access, slow for insert/delete in middle.

# Mini Project 1 — Simple Data Analytics Toolkit
Real-World Use

Companies analyze numbers (sales, temperatures, stock prices, etc.)

We’ll build a small analytics tool using arrays.

## Features

Add data

Remove data

Search value

Find max / min

Average

Moving average (important in finance)

```python
# Mini Project 1 — Simple Data Analytics Toolkit
class DataAnalytics:
    def __init__(self):
        self.data = []

    def add(self, value):
        self.data.append(value)

    def remove(self, value):
        if value in self.data:
            self.data.remove(value)

    def search(self, value):
        return value in self.data

    def maximum(self):
        return max(self.data) if self.data else None

    def minimum(self):
        return min(self.data) if self.data else None

    def average(self):
        return sum(self.data) / len(self.data) if self.data else 0

    def moving_average(self, k):
        result = []
        for i in range(len(self.data) - k + 1):
            window = self.data[i:i+k]
            result.append(sum(window) / k)
        return result
```

```python
tool = DataAnalytics()

tool.add(10)
tool.add(20)
tool.add(30)
tool.add(40)
tool.add(50)

print("Data:", tool.data)
print("Max:", tool.maximum())
print("Min:", tool.minimum())
print("Avg:", tool.average())
print("Moving Avg (k=3):", tool.moving_average(3))
```

```python
tool.remove(300)
print("Data after removal:", tool.data)
tool.remove(30)
print("Data after removal 30:", tool.data)
```

#  List Methods & Time Complexity

```python
# Append: O(1) amortized
fruits.append("orange")
print(fruits)  # ['apple', 'blueberry', 'cherry', 'orange']

# Insert: O(n) - elements must be shifted
fruits.insert(1, "mango")
print(fruits)  # ['apple', 'mango', 'blueberry', 'cherry', 'orange']

# Remove: O(n) - must search and shift
fruits.remove("blueberry")
print(fruits)  # ['apple', 'mango', 'cherry', 'orange']

# Pop: O(1) for last, O(n) for arbitrary index
last = fruits.pop()
print(last)    # orange
print(fruits)  # ['apple', 'mango', 'cherry']

# Index: O(n)
print(fruits.index("cherry"))  # 2

# Sort: O(n log n)
numbers = [3, 1, 4, 1, 5, 9, 2]
numbers.sort()
print(numbers)  # [1, 1, 2, 3, 4, 5, 9]

# List comprehension (powerful feature)
squares = [x**2 for x in range(10)]
print(squares)  # [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
```
